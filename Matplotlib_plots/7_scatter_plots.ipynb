```python
# Trending YouTube Videos Data extracted through API
data = pd.read_csv("c:\\Users\\shaba\\PycharmProjects\\Data-Analysis\\Data\\2019-05-31-data.csv")
data.head()
```

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.style.use('seaborn-v0_8')

plt.scatter(view_count, likes,c= ratio,cmap = 'summer',
 edgecolor = 'black', alpha= 0.75)

c_bar = plt.colorbar()
c_bar.set_label("Like/Dislike Ratio")

plt.xscale('log')
plt.yscale('log')

plt.title("Trending YouTube Views")
plt.xlabel("View Count")
plt.ylabel("Total Likes")
```
