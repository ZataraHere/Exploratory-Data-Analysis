```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Note: If you have more than 5 labels it is reccomended to use bars

```python
data = [20,60, 40, 100]
labels = ['Two', 'Six','Four','Ten']
color = ["#008ed59b", "#22b9c4a4", "#2fdf1f9b", "#6417f39d"]
plt.pie(data,labels=labels, colors= color, wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Piechart")
```

```python
language = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#']
popularity = [59219, 55466, 47544, 36443, 35917, 31991, 27097]
explode = [0,0,0,0.1,0,0,0]

plt.pie(popularity,labels=language, explode=explode, shadow=True, 
        startangle=90, autopct='%1.1f%%',
         wedgeprops={'edgecolor': 'black'})

plt.title("Most Popular Languages")
```
