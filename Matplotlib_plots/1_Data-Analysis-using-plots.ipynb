# Simple plots

```python
ages_x = np.arange(10,20)

dev_x = np.random.randint(1,100,10)
dev_y = np.random.randint(1,100,10)

plt.plot(ages_x, dev_x,linestyle = '--',marker = '.', color = 'blue', label = 'x-plot')

plt.plot(ages_x, dev_y, color = 'green',marker = 'o', label = 'y-plot')

plt.legend()
plt.title('x vs y ')
plt.ylabel('x,y')
plt.xlabel('age')

plt.tight_layout()
plt.grid()
```

```python
plt.style.use('fivethirtyeight')

ages_x = np.arange(10,20)

dev_x = np.random.randint(1,100,10)
dev_y = np.random.randint(1,100,10)

plt.plot(ages_x, dev_x, label = 'x-plot')

plt.plot(ages_x, dev_y, label = 'y-plot')

plt.legend()
plt.title('x vs y ')
plt.ylabel('x,y')
plt.xlabel('age')
```
