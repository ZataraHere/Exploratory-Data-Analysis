```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = sns.load_dataset('flights')
```

```python
x = data.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.clustermap(x)
```

```python
# Normalizing values
sns.clustermap(x, metric='correlation',z_score=0, annot= True)
```

```python
sns.clustermap(x, metric='correlation',z_score=0, row_cluster=False, annot= True)
```
