#### Box-Plots

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
```

```python
# Univariate

sns.boxplot(data['total_bill'])
```

```python
sns.catplot(x= 'day', y = 'total_bill', kind='box', data=data)
```

```python
sns.catplot(x= 'day', y = 'total_bill', kind='box',hue='sex', data=data)
```
