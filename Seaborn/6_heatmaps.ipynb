```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = sns.load_dataset('flights')
```

```python
data.head()
```

```python
x = data.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar=False)
```

```python
sns.heatmap(x, linewidths=0.5)
```

```python
sns.heatmap(x, linewidths=0.5,annot=True, fmt='d')
```

```python
sns.heatmap(x, linewidths=0.5, annot=True, fmt = 'd', cmap = 'summer')
```
