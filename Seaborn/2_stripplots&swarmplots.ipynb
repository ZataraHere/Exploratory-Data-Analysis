#### Categorical Scatter Plots

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.catplot(x ='day',y = 'tip',kind = 'strip', data = data )
#jitter for adding noise, it has a default value
```

```python
sns.catplot(x ='day',y = 'tip',kind = 'swarm', data = data )
```

```python
sns.catplot(x ='day',y = 'tip',kind = 'swarm',hue = 'sex', data = data )
```

```python
sns.swarmplot(x = 'day', y = 'tip', data = data)
```

```python
colors = ['red', 'green', 'blue', 'yellow']
sns.catplot(x ='day',y = 'tip',kind = 'strip', data = data , palette=colors)
```
